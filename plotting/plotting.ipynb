```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('player_scores_sortedMutate.csv')
```

```python
def Graph(df, title="Player Scores over Generations"):
    x = df['Generation']

    for i in range(1, len(df.columns)):
        plt.plot(x, df[df.columns[i]], label=df.columns[i])

    # Add axis labels and legend to the plot
    plt.xlabel('Generation')
    plt.ylabel('Score (%)')
    plt.title(title)
    plt.legend()
```

```python
df = pd.read_csv('player_scores_sortedMutate.csv')
Graph(df, "Sorted Mutate")
```
